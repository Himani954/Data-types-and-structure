# Q6. Explain the concept of generators in Python and how they are defined.

# Ans6. Generators in python

A genertor is a special type of iterable in Python that generates values on-the-fly,rather that storing them in memory all at once. Generators are useful for handling large datasets or creating sequences that would be too memory- intensive to store in a list.

Defining generators

Generators are defined using function with the yield keyword. When a generator function is called, it returns a generator object, which can be iterated over using a for loop or the next() function.

Example

```python
python code
def infinite_sequence():
  num = 0
  while True:
    yield num
    num +=1

gen = infinite_sequence()
print(next(gen)) #Output: 0
print(next(gen)) #Output: 1
print(next(gen)) #Output: 2
```

In this example, the

infinite_sequence function is a generator that yields an infinite sequence of numbers.The yield keyword is used to produce a value and suspend the function's execution until the next value is requested.

key Benefits

* Memory efficiency: Generators use significantly less memory than storing large datasets in lists.

* Flexibility: Generators can be used to create complex sequences or handle large datasets that wouldn't fit in memory.

Generator Expression

Generators can also be defined using generator expressions, which are similar to list comprehension but use parentheses instead of brackets:

```python
gen_expr = (x**2 for x in range (10))
for num in gen_expr:
  print(num)
```

This generators expression generates the squares of numbers from 0 to 9.

# Q7. What are the advantages of using generators over regular functions?

# Ans7. Advantages of Genertors

1. Memory Efficiency: Generators use signicantly less memory than regular fuctions that return large lists or datasets.

2. Lazy Evalution: Generators only generate values when needed, reducing unnecessary computations.

3. Flexibility: Generators can be used to create complex sequences or handle large datasets that wouldn't fit in memory.

4. Improved performance: Generators can improve performance by avoiding the need to create and store large datesets in memory.

5. Infinite Sequences: Generators can be used to create infinite sequences, which would be impossible with regular functions.

Use Cases

1. Handling Large Datasets: Generators are ideal for handling large datasets that don't fit in memory.

2. Creating Complex Sequences: Generators can be used to create complex sequences, such as Fibonacci numbers or prime numbers.

3. Real- time Data processing:
Generators can be used to process real-time data, such as streaming data from a file or network.

```python
# Regular fuction
def squares(n):
  result = []
  for i in range(n):
    result.append(i**2)
  return result

#Generator
def squares_generator(n):
  for i in range(n):
    yield i**2


# Memory usage comparison
import sys
n = 1000000
print(sys.getsizeof(squares(n))) # output: large memory usage
print(sys.getsizeof(squares_generator(n))) # output: small memory usage
```

In this example, the squares fuction returns a large list, while the squares_gen generator uses singificantly less memory.

# Q8. What is a lambda function in Python and when is it typically used?

# ANS8. Lambda Fuction in Python

A lambda function is a small, anonymous function in Python that can take any number of arguments, but can only have one expression. It's defined using the lambda keyword.

Syntax

```python
python code

add = lambda x, y: x + y
print(add(3,4)) # Output: 7
```

```python
add = lambda x, y: x + y
print(add(3, 4)) # Output: 7
```

Typical Use Cases

1. One-time use: Lambda fuctions are useful when you need a small, one-time use function.

2. Higher-order functions: Lambda functions are often used as arguments to higher-order as arguments to higher-order functions, such as map(), filter(), and reduce().

3. Data processing: Lambda functions can be used to process data in a concise way.

Examples of Use Cases

```python
# Using lambda with map()
numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x**2, numbers))
print(squares) # Output: [1, 4, 9, 16, 25]

# Using lambda with filter()
numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers) # Output: [2, 4]

#  Using lambda with sorted()
students = [{'name': 'john', 'age':20}, {'name':'Alice', 'age': 19}]
students_sorted = sorted(students, key=lambda x: x['age'])
print(students_sorted) # Output: [{'name': 'Alice', 'age': 19}, {'name': 'john', 'age': 20}]
```

# Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

# Ans10. Map, Reduce, and Filter Functions in Python

These three functions are used to process iterables in Python,but they serve different purpose:

Map()
* Purpose: Applies a function to each item in an iterable and returns a new iterable with the results.
* Output: A new iterable with the same number of items as the input iterable.

Example:

```python
numbers = [1, 2, 3, 4, 5]
doubled = map(lambda x: x * 2, numbers)
print(list(doubled)) # Output: [2, 4, 6, 8, 10]
```

Filter()
* Purpose: Selects items from an iterable that meet a certain condition and returns a new iterable with only those items.
* Output: A new iterable with a subset of the items from the input iterable.

Example:

```python
numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers) # Output: [2, 4]
```

Reduce()
* Purpose: Applies a function to the items in an iterable, reducing them to single output value.
*Output: A single value that is the result of applying the function to all items in the iterable.

Example:

```python
import functools
numbers = [1, 2, 3, 4, 5]
sum_numbers = functools.reduce(lambda x, y: x + y, numbers)
print(sum_numbers) # Output: 15
```

Key Differences
* Map: Transforms each item in an iterable.
* Filter: Selects a subset of items from an iterable.
* Reduce: Reduces an iterable to a single output value.

These functions are fundamental building blocks for functional programming in python, and understanding their differences is essential for using them effectively.

# Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

# Example usage:
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```
