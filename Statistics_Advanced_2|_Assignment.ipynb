# **Question 1: What is hypothesis testing in statistics?**

# **Ans1.**
# Hypothesis Testing in Statistics
Hypothesis testing is a statistical technique used to make inferences about a population parameter based on a sample of data. It involves formulating a hypothesis about the population parameter and then testing it using statistical methods.

Steps in Hypothesis Testing
The steps involved in hypothesis testing are:

1. Formulate a Null Hypothesis : Specify a null hypothesis that represents a statement of no effect or no difference.
2. Formulate an Alternative Hypothesis : Specify an alternative hypothesis that represents a statement of an effect or difference.
3. Collect Sample Data : Collect a sample of data from the population.
4. Calculate a Test Statistic : Calculate a test statistic that measures the difference between the sample data and the null hypothesis.
5. Determine the P-Value : Determine the p-value, which represents the probability of observing the test statistic (or a more extreme value) assuming that the null hypothesis is true.
6. Make a Decision : Make a decision to reject or not reject the null hypothesis based on the p-value and a predetermined significance level.

Purpose of Hypothesis Testing
The purpose of hypothesis testing is to:

- Test Hypotheses : Test hypotheses about population parameters based on sample data.
- Make Inferences : Make inferences about the population based on the sample data.
- Determine Significance : Determine whether the results are statistically significant.

Importance of Hypothesis Testing
Hypothesis testing is important because:

- Statistical Inference : It enables researchers to make inferences about populations based on sample data.
- Evidence-Based Decision Making : It provides a framework for making evidence-based decisions.
- Scientific Research : It is a fundamental tool in scientific research, allowing researchers to test hypotheses and draw conclusions about the world.

By understanding hypothesis testing, researchers can apply it to various statistical problems and make informed decisions based on data analysis.



# **Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?**

# **Ans2.**
# Null Hypothesis
The null hypothesis is a statement of no effect or no difference, typically denoted as H0. It represents a default position that there is no significant relationship or difference between variables.

Alternative Hypothesis
The alternative hypothesis is a statement of an effect or difference, typically denoted as H1 or Ha. It represents a statement that there is a significant relationship or difference between variables.

Key Differences
The key differences between the null hypothesis and the alternative hypothesis are:

- Direction of the Hypothesis : The null hypothesis states that there is no effect or difference, while the alternative hypothesis states that there is an effect or difference.
- Purpose : The null hypothesis is used as a baseline to test against, while the alternative hypothesis is the statement that the researcher is trying to support or demonstrate.

Example
For example, if a researcher wants to test whether a new medication is effective in lowering blood pressure:

- Null Hypothesis (H0) : The new medication has no effect on blood pressure.
- Alternative Hypothesis (H1) : The new medication has an effect on blood pressure.

By understanding the null and alternative hypotheses, researchers can design studies and interpret results more effectively.



# **Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

# **Ans3.**
# Significance Level
The significance level, denoted as α (alpha), is the maximum probability of rejecting a true null hypothesis. It's a threshold value that determines whether the results of a test are statistically significant.

Role in Hypothesis Testing
The significance level plays a crucial role in deciding the output of a test:

1. Determining Statistical Significance : If the p-value is less than the significance level (α), the results are considered statistically significant, and the null hypothesis is rejected.
2. Type I Error Control : The significance level controls the probability of Type I errors (rejecting a true null hypothesis). A smaller α reduces the risk of Type I errors.
3. Decision Making : The significance level helps researchers make decisions about the null hypothesis. If the p-value is less than α, the null hypothesis is rejected; otherwise, it's not rejected.

Common Significance Levels
Commonly used significance levels are:

- α = 0.05: A widely used significance level, indicating a 5% chance of rejecting a true null hypothesis.
- α = 0.01: A more stringent significance level, indicating a 1% chance of rejecting a true null hypothesis.

Importance
The significance level is important because it:

- Provides a Framework : Provides a framework for making decisions about the null hypothesis.
- Controls Error Rates : Controls the probability of Type I errors.
- Helps with Interpretation : Helps researchers interpret the results of a test and make informed decisions.

By understanding the significance level, researchers can design studies and interpret results more effectively.



# **Question 4: What are Type I and Type II errors? Give examples of each.**


# **Ans4.**
# Type I Error
A Type I error occurs when a true null hypothesis is rejected. This is also known as a "false positive" finding.

Example of Type I Error
Suppose a court is trying to determine whether a defendant is guilty of a crime. The null hypothesis is that the defendant is innocent.

- Null Hypothesis (H0) : The defendant is innocent.
- Type I Error : The court rejects the null hypothesis and convicts the defendant, even though they are actually innocent.

Type II Error
A Type II error occurs when a false null hypothesis is not rejected. This is also known as a "false negative" finding.

Example of Type II Error
Using the same court example:

- Null Hypothesis (H0) : The defendant is innocent.
- Type II Error : The court fails to reject the null hypothesis and acquits the defendant, even though they are actually guilty.

More Examples
- Medical Testing :
    - Type I Error : A test indicates that a person has a disease when they actually don't (false positive).
    - Type II Error : A test fails to detect a disease in a person who actually has it (false negative).
- Quality Control :
    - Type I Error : A quality control test indicates that a batch of products is defective when it's actually not (false positive).
    - Type II Error : A quality control test fails to detect defects in a batch of products that are actually defective (false negative).

By understanding Type I and Type II errors, researchers and decision-makers can design studies and make decisions more effectively.



# **Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.**



# **Ans5.**
# Z-Test vs T-Test
Both Z-tests and T-tests are used for hypothesis testing, but they differ in their assumptions and applications.

Z-Test
A Z-test is used when:

- Population Standard Deviation is Known : The population standard deviation (σ) is known.
- Large Sample Size : The sample size is large (typically n ≥ 30).
- Normal Distribution : The population distribution is normal or approximately normal.

T-Test
A T-test is used when:

- Population Standard Deviation is Unknown : The population standard deviation (σ) is unknown.
- Small Sample Size : The sample size is small (typically n < 30).
- Normal Distribution : The population distribution is normal or approximately normal.

Key Differences
The key differences between Z-tests and T-tests are:

- Known vs Unknown Population Standard Deviation : Z-tests require a known population standard deviation, while T-tests do not.
- Sample Size : Z-tests are typically used with large sample sizes, while T-tests are used with small sample sizes.

When to Use Each
- Use a Z-Test :
    - When comparing a sample mean to a known population mean and the population standard deviation is known.
    - When working with large sample sizes and the population distribution is normal or approximately normal.
- Use a T-Test :
    - When comparing a sample mean to a known population mean and the population standard deviation is unknown.
    - When working with small sample sizes and the population distribution is normal or approximately normal.

By understanding the differences between Z-tests and T-tests, researchers can choose the most appropriate statistical test for their research question.



# **Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.**


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10     # number of trials
p = 0.5    # probability of success
size = 1000  # number of samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=np.arange(n+2) - 0.5, edgecolor='black', density=True)
plt.title('Binomial Distribution Histogram (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(range(n+1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# **Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.**

# **sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,**

# **50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,**

# **50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,**

# **50.3, 50.4, 50.0, 49.7, 50.5, 49.9]**


```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
mu_0 = 50        # Hypothesized population mean
sigma = 0.5      # Assumed known population standard deviation
alpha = 0.05     # Significance level

# Sample statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Z-statistic
z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z)))

# Output
print(f"Sample mean = {sample_mean:.4f}")
print(f"Z-statistic = {z:.4f}")
print(f"P-value = {p_value:.4f}")

# Decision
if p_value < alpha:
    print("Reject the null hypothesis (H₀).")
else:
    print("Fail to reject the null hypothesis (H₀).")
```

# **Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
mu, sigma = 100, 15     # mean and standard deviation
n = 100                 # sample size
data = np.random.normal(mu, sigma, n)

# Step 2: Calculate 95% Confidence Interval
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
conf_level = 0.95
alpha = 1 - conf_level

# t critical value (since sigma is unknown, we use t-distribution)
t_crit = stats.t.ppf(1 - alpha/2, df=n-1)
margin_of_error = t_crit * (sample_std / np.sqrt(n))

ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 3: Plot histogram with confidence interval
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='blue', linestyle='--', label='Sample Mean')
plt.axvline(ci_lower, color='red', linestyle='--', label='95% CI Lower')
plt.axvline(ci_upper, color='green', linestyle='--', label='95% CI Upper')

plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# **Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.**



```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = (data - mean) / std
    return z_scores

# Simulate dataset
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=100)

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Display a few Z-scores
print("First 5 Z-scores:", np.round(z_scores[:5], 2))

# Plot histogram of standardized data (Z-scores)
plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7, color='skyblue')
plt.title('Histogram of Z-scores (Standardized Data)')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.axvline(0, color='red', linestyle='--', label='Mean (Z = 0)')
plt.legend()
plt.show()
```
